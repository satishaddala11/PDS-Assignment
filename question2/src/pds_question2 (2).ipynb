```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset from CSV file
df = pd.read_csv("sample_data/StudentsPerformance.csv", encoding="ISO-8859-1")
 # Replace with the actual file name
print(df.info())
```

```python
# Removing rows with null values
df_cleaned = df.dropna()
df_cleaned.info()
```

```python
df_cleaned.to_csv('clean_data2.csv',index=False)
```

```python
# Performing five different data visualizations on the student performance dataset

# Visualization 1: Distribution of Math Scores
plt.figure(figsize=(8,5))
sns.histplot(df_cleaned["math score"], bins=10, kde=True, color="blue")
plt.xlabel("Math Score")
plt.ylabel("Frequency")
plt.title("Distribution of Math Scores")
plt.show()

# Visualization 2: Boxplot of Scores by Gender
plt.figure(figsize=(8,5))
sns.boxplot(x=df_cleaned["gender"], y=df["math score"])
plt.xlabel("Gender")
plt.ylabel("Math Score")
plt.title("Math Scores by Gender")
plt.show()

# Visualization 3: Relationship Between Reading and Writing Scores
plt.figure(figsize=(8,5))
sns.scatterplot(x=df_cleaned["reading score"], y=df["writing score"], hue=df["gender"], style=df["gender"])
plt.xlabel("Reading Score")
plt.ylabel("Writing Score")
plt.title("Reading vs. Writing Scores by Gender")
plt.show()

# Visualization 4: Impact of Lunch Type on Math Scores
plt.figure(figsize=(8,5))
sns.boxplot(x=df_cleaned["lunch"], y=df["math score"])
plt.xlabel("Lunch Type")
plt.ylabel("Math Score")
plt.title("Math Scores by Lunch Type")
plt.show()

# Visualization 5: Test Preparation Course and Average Scores
plt.figure(figsize=(8,5))
df["average score"] = (df_cleaned["math score"] + df["reading score"] + df["writing score"]) / 3
sns.barplot(x=df["test preparation course"], y=df["average score"], palette="viridis")
plt.xlabel("Test Preparation Course")
plt.ylabel("Average Score")
plt.title("Average Scores by Test Preparation Course")
plt.show()
```

1. Distribution of Math Scores (Histogram)
Easier Analysis:

Helps understand the overall distribution of math scores.
Identifies if scores follow a normal distribution, are skewed, or have outliers.
Useful for understanding the general performance level of students in math.
2. Math Scores by Gender (Boxplot)
Easier Analysis:

Allows easy comparison of math performance between male and female students.
Highlights median scores and possible outliers for each gender.
Helps identify any gender-based performance differences in math.
3. Relationship Between Reading and Writing Scores (Scatter Plot)
Easier Analysis:

Shows the correlation between reading and writing performance.
Helps in identifying whether students who score well in one subject also perform well in another.
Useful for predicting writing scores based on reading scores.
4. Math Scores by Lunch Type (Boxplot)
Easier Analysis:

Investigates whether students receiving free/reduced lunch perform differently from those with standard lunch.
Highlights possible socioeconomic factors affecting academic performance.
Helps in understanding if students from lower-income backgrounds face additional learning challenges.
5. Impact of Test Preparation Course on Scores (Bar Chart)
Easier Analysis:

Compares average scores of students who took a test preparation course versus those who did not.
Evaluates the effectiveness of test preparation programs.
Helps students and educators decide whether test prep courses provide a significant advantage.

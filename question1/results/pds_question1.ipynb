```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset from CSV file
df = pd.read_csv("sample_data/clean_data.csv", encoding="ISO-8859-1")
 # Replace with the actual file name
print(df.info())
```

```python
 # Visualization 1: Distribution of Grip Strength
plt.figure(figsize=(8,5))
sns.histplot(df["Grip strength"], bins=6, kde=True)
plt.xlabel("Grip Strength (kg)")
plt.ylabel("Frequency")
plt.title("Distribution of Grip Strength")
plt.show()




# Visualization 2: Boxplot of Grip Strength based on Frailty
plt.figure(figsize=(8,5))
sns.boxplot(x=df["Frailty"], y=df["Grip strength"])
plt.xlabel("Frailty")
plt.ylabel("Grip Strength (kg)")
plt.title("Grip Strength Distribution by Frailty")
plt.show()


```
